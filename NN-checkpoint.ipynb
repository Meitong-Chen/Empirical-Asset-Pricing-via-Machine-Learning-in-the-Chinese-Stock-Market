```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号 #有中文出现的情况，需要u'内容'
from sklearn.metrics import r2_score
```

```python
def roos(y_pre,y_true):
    up=np.sum((y_true-y_pre)**2)
    down=np.sum(y_true**2)
    return 1-up/down
```

```python
df=pd.read_excel('data.xlsx')
```

```python
from sklearn.model_selection import train_test_split 
df_train,df_test=train_test_split (df,test_size=0.3,random_state=123)
```

```python
df_train
```

```python
df_test
```

```python
X_train=df_train[df_train.columns[1:]]   #训练集输入
y_train=np.array(df_train['return']).flatten()  #训练集输出
X_test=df_test[df_test.columns[1:]]  #测试集输入  
y_test=np.array(df_test['return']).flatten()  #测试集输出
```

```python
#标准化
from sklearn.preprocessing import StandardScaler
scale=StandardScaler() #将输入数据进行标准化
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
```

```python
#随机森林
#调参
import sklearn
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
estimator=RandomForestRegressor(random_state=1)
param_grid={'n_estimators':range(5,30,5),'max_depth':range(5,30,5)}
grid=sklearn.model_selection.GridSearchCV(estimator, param_grid, scoring='neg_mean_squared_error',cv=5)
grid.fit(X_train, y_train)
grid.best_params_
```

```python
rf = RandomForestRegressor(max_depth= 10, n_estimators=10,random_state=0)
rf.fit(X_train, y_train)
y_pred_rf=rf.predict(X_test)
mse=mean_squared_error(y_pred_rf,y_test)
mae=mean_absolute_error(y_pred_rf,y_test)
rmse=mse**0.5
roos2=r2_score(y_pred_rf,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
rf_res=[mse,rmse,mae,roos2]
```

```python
#画图
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_rf,label='predict by rf') #模型预测值图
plt.legend()
```

```python
import seaborn as sns
importances = pd.Series(data=rf.feature_importances_,index=df_train.columns[1:]).sort_values(ascending=False)
plt.figure(figsize=(8,4))
sns.barplot(y=importances.index,x=importances.values,orient='h')
plt.xlabel('feature importance')
plt.ylabel('feature')
plt.savefig('随机森林重要性系数.jpg',dpi=600,bbox_inches='tight')
```

```python
#GBDT调参
estimator=GradientBoostingRegressor(random_state=1)
param_grid={'n_estimators':range(100,600,100),'learning_rate':[1e-4,1e-3,1e-2,1e-1,1]}
grid=sklearn.model_selection.GridSearchCV(estimator, param_grid, scoring='neg_mean_squared_error',cv=5)
grid.fit(X_train,y_train)
grid.best_params_
```

```python
gbdt=GradientBoostingRegressor(n_estimators=500,learning_rate=0.1,random_state=1)
gbdt.fit(X_train,y_train)
y_pred_gbdt=gbdt.predict(X_test)
mse=mean_squared_error(y_pred_gbdt,y_test)
mae=mean_absolute_error(y_pred_gbdt,y_test)
rmse=mse**0.5
roos2=roos(y_pred_gbdt,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
gbdt_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_gbdt,label='predict by gbdt') #模型预测值图
plt.legend()
```

```python
importances = pd.Series(data=gbdt.feature_importances_,index=df_train.columns[1:]).sort_values(ascending=False)
plt.figure(figsize=(8,4))
sns.barplot(y=importances.index,x=importances.values,orient='h')
plt.xlabel('feature importance')
plt.ylabel('feature')
plt.savefig('GBDT重要性系数.jpg',dpi=600,bbox_inches='tight')
```

```python
#神经网络调参
from sklearn.neural_network import MLPRegressor
estimator=MLPRegressor(hidden_layer_sizes=(256,128,64,32), activation='relu',solver='adam',alpha=0.1,max_iter=2000,random_state=12)
param_grid={'learning_rate_init':[1e-3,1e-2]}
grid=sklearn.model_selection.GridSearchCV(estimator, param_grid, scoring='neg_mean_squared_error',cv=5)
grid.fit(X_train,y_train)
grid.best_params_
```

```python
mlp=MLPRegressor(hidden_layer_sizes=(256,128,64,32), activation='relu',solver='adam',alpha=0.1,max_iter=5000,learning_rate_init=0.01,random_state=12)
mlp.fit(X_train,y_train)
y_pred_mlp=mlp.predict(X_test)
mse=mean_squared_error(y_pred_mlp,y_test)
mae=mean_absolute_error(y_pred_mlp,y_test)
rmse=mse**0.5
roos2=roos(y_pred_mlp,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
mlp_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_mlp,label='predict by mlp') #模型预测值图
plt.legend()
```

```python
#pls
from sklearn.cross_decomposition import PLSRegression
pls=PLSRegression()
pls.fit(X_train,y_train)
y_pred_pls=pls.predict(X_test)
mse=mean_squared_error(y_pred_pls,y_test)
mae=mean_absolute_error(y_pred_pls,y_test)
rmse=mse**0.5
roos2=roos(y_pred_pls,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
pls_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_pls,label='predict by pls') #模型预测值图
plt.legend()
```

```python
#弹性网络
from sklearn.linear_model import ElasticNet
estimator=ElasticNet()
param_grid={'l1_ratio':[.1,.5,.7,.9,.99,1],'alpha':[.1,.5,.7,.9,.99,1]}
grid=sklearn.model_selection.GridSearchCV(estimator, param_grid, scoring='neg_mean_squared_error',cv=5)
grid.fit(X_train,y_train)
grid.best_params_
```

```python
elnet=ElasticNet(l1_ratio=0.7,alpha=0.5)
elnet.fit(X_train,y_train)
y_pred_elnet=elnet.predict(X_test)
mse=mean_squared_error(y_pred_elnet,y_test)
mae=mean_absolute_error(y_pred_elnet,y_test)
rmse=mse**0.5
roos2=roos(y_pred_elnet,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
elnet_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_elnet,label='predict by elnet') #模型预测值图
plt.legend()
```

```python
#GLS
import statsmodels.api as sm
X_train_g = sm.add_constant(X_train)  #构造常数项
X_test_g = sm.add_constant(X_test) 
gls = sm.GLS(y_train,X_train_g).fit()
print(gls.summary())
y_pred_gls=gls.predict(X_test_g)
mse=mean_squared_error(y_pred_gls,y_test)
mae=mean_absolute_error(y_pred_gls,y_test)
rmse=mse**0.5
roos2=roos(y_pred_gls,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
gls_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_gls,label='predict by gls') #模型预测值图
plt.legend()
```

```python
#OLS
X_train_o = sm.add_constant(X_train)  #构造常数项
X_test_o = sm.add_constant(X_test) 
ols = sm.OLS(y_train,X_train_o).fit()
print(ols.summary())
y_pred_ols=ols.predict(X_test_o)
mse=mean_squared_error(y_pred_ols,y_test)
mae=mean_absolute_error(y_pred_ols,y_test)
rmse=mse**0.5
roos2=roos(y_pred_ols,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
ols_res=[mse,rmse,mae,roos2]
```

```python
#画图
plt.figure(figsize=(12,4))
plt.plot(y_test,label='true')  #真实值画图
plt.plot(y_pred_ols,label='predict by gls') #模型预测值图
plt.legend()
```

```python
#PCA+OLS
from sklearn.decomposition import PCA
pca=PCA(n_components=0.8)
X_train_p=pca.fit_transform(X_train)
X_test_p=pca.transform(X_test)
X_train_o = sm.add_constant(X_train_p)  #构造常数项
X_test_o = sm.add_constant(X_test_p) 
pols = sm.OLS(y_train,X_train_o).fit()
print(pols.summary())
y_pred_pols=pols.predict(X_test_o)
mse=mean_squared_error(y_pred_pols,y_test)
mae=mean_absolute_error(y_pred_pols,y_test)
rmse=mse**0.5
roos2=roos(y_pred_pols,y_test)
print('mse:{},rmse:{},mae:{},roos2:{}'.format(mse,rmse,mae,roos2))
pols_res=[mse,rmse,mae,roos2]
```

```python
res=pd.DataFrame([rf_res,gbdt_res,mlp_res,pls_res,elnet_res,gls_res,ols_res,pols_res])
res.columns=['mse','rmse','mae','roos2']
res.index=['随机森林','梯度提升树','神经网络','PLS','弹性网络','广义线性回归','普通线性回归','PCA+OLS']
res
```
